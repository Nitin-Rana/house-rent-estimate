```python
import pandas as pd
```

```python
df = pd.read_csv("houses_to_rent_2.csv")
```

```python
df.head(10)
```

```python
df.loc[df['floor'] == '-', 'floor'] = 0
df['floor'] = df['floor'].astype('int64')

df['Hassoc_tax'] = df['hoa (R$)']

df['rent'] = df['rent amount (R$)']

df['prop_tax'] = df['property tax (R$)']

df['fire_insur'] = df['fire insurance (R$)']
```

```python
df.head(10)
```

```python
df.columns
```

```python
df_out = df[['city', 'area', 'rooms', 'bathroom', 'parking spaces', 'floor', 'animal', 'furniture', 'Hassoc_tax', 'rent', 'prop_tax', 'fire_insur']]
```
